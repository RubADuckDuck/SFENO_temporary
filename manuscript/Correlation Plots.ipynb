```python
import os
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

pred_dir_name = 'predictions'
pred_dir_name = os.path.join('predictions', '40_0')

dataset_dir_name_ls = os.listdir(pred_dir_name)

result_relative_path = os.path.join('sfeno_data','results')

exp_name_ls = ['epoch_4000_net_info_False_0', 'epoch_4000_net_info_True_0']

pred_nm = 'pred.npy'
targ_nm = 'targ.npy'

save_dir_path = 'pred_plots'
```

```python
# for box plot 0 dim: data 1 dim: exp
bp_2d = [[0] * len(exp_name_ls)] * len(dataset_dir_name_ls)
debug2d = [[0] * len(exp_name_ls)] * len(dataset_dir_name_ls)
```

```python
for idx_data, dataset_nm in enumerate(dataset_dir_name_ls):
    for idx_exp, exp_nm in enumerate(exp_name_ls):
        current_file_path = os.path.join(
            pred_dir_name,
            dataset_nm,
            result_relative_path,
            exp_nm,
        )
        
        cur_pred_path = os.path.join(current_file_path, pred_nm)
        cur_targ_path = os.path.join(current_file_path, targ_nm)
        cur_prefix = f'{dataset_nm}_{exp_nm}_'
        
        print('-----------------------')
        print(current_file_path)
        print(cur_pred_path)
        print(cur_targ_path)
        
        pred_arr = np.load(cur_pred_path)
        idx_remove = np.where((-100 > pred_arr) | (pred_arr > 100))
        targ_arr = np.load(cur_targ_path)
        
        pred_arr = np.delete(pred_arr, idx_remove)
        targ_arr = np.delete(targ_arr, idx_remove)
        
        max_plt = np.max(pred_arr) + 1
        min_plt = np.min(targ_arr) - 1
        scale = max_plt - min_plt
        
        cur_corr = np.corrcoef(pred_arr,targ_arr)[0,1]
        cur_corr = round(cur_corr,5)
        print(cur_corr)
        
        plt.figure(figsize=(6,6))
        
        plt.scatter(pred_arr, targ_arr, alpha=0.5)
        plt.xlim([min_plt, max_plt])
        plt.ylim([min_plt, max_plt])
        plt.xlabel('Prediction')
        plt.ylabel('Target')
        
        _,epoch,_,_,net_info,_ = exp_nm.split('_')

        cur_plt_title = f'{dataset_nm}\nnetwork constraints: {net_info}\nepoch: 4000 batch: 16'

        plt.title(cur_plt_title, loc='left', fontsize = 26)
        plt.text(min_plt+0.03*scale, max_plt-0.1*scale, f'correlation: {cur_corr}', fontsize = 26,
                bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 5})
        
        cur_plt_save_path = get_valid_file_nm(os.path.join(save_dir_path, cur_prefix))
        print(f'Saving to {cur_plt_save_path}')
        plt.tight_layout()
        
        plt.savefig(cur_plt_save_path, dpi=300)
        
        plt.show()
```
