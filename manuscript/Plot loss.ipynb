```python
import pandas as pd
import numpy as np
```

```python
metric_df = pd.read_csv('metrics_molineli.csv')
```

```python
validation = metric_df[['epoch','val_loss']].dropna().set_index('epoch')
train = metric_df[['epoch','train_loss']].dropna().set_index('epoch')
```

```python
validation
```

```python

from matplotlib.pylab import plt
from numpy import arange
 
start = 0
end = 4000
name = f'molineli_{start}_{end}'
    
# Retrieve each dictionary's values
train_values = train['train_loss'][:end]
val_values = validation['val_loss'][:end]
 
# Generate a sequence of integers to represent the epoch numbers
epochs = range(start, end)
 
# Plot and label the training and validation loss values
plt.plot(epochs, train_values, label='Training Loss',marker='D', alpha=0.5, lw=0.5, ms=3)
plt.plot(epochs, val_values, label='Validation Loss',marker='^', alpha=0.5, lw=0.5, ms=3)
 
# Add in a title and axes labels
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
 
 
# Display the plot
plt.legend(loc='best')
plt.savefig("%s.png"%(name))
plt.show()
```
